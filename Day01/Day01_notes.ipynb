##  âœ… Dynamic Typing

Python decides the type of a variable at runtime, not in advance.

Example:
x = 10      # x is an integer
x = "Hello" # now x becomes a string


ðŸ“Œ The type of the variable can change during execution.

```python
# Dynamic Typing  : python automaticlly understand which type of data it is , so python support Dynamic typing
# C , C++ support static typing
# ex
a = 5
b = 5.5
print(type(a))
print(type(b))
```

## âœ… Dynamic Binding

Python decides which variable's value to use at runtime, based on the current assignment.

def show(value):
    print(value)

x = "Hello"

y = 100

show(x)      # prints: Hello

show(y)      # prints: 100


ðŸ“Œ The function call binds to the variable's current value at runtime.

```python
# Dynamic Binding
a = 5
print(a)
a = 'Akash'
print(a)
```
