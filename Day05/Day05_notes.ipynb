```python
fruits = ["apple", "banana", "mango"]
print(fruits)
```

```python
numbers = [10, 20, 30, 40]
print(numbers[0])   # first element
print(numbers[-1])  # last element
```

```python
colors = ["red", "blue"]
colors.append("green")       # add at end
colors.insert(0, "yellow")   # add at index 0
print(colors)
```

```python
pets = ["cat", "dog", "parrot"]
pets[1] = "rabbit"
print(pets)
```

```python
cities = ["Delhi", "Mumbai", "Kolkata", "Chennai"]
cities.remove("Mumbai")   # remove by value
cities.pop(0)             # remove by index
print(cities)
```

```python
languages = ["Python", "Java", "C++"]
for lang in languages:
    print(lang)
```
