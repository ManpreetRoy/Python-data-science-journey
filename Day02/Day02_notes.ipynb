````python
# Arithmetic
# Relational
# Logical
# Bitwise
# Assignment
# Membership
````

````python
# Arithmetic Operator
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
````

````python
#Relational Operator
print(4>5)

print(4<5)

print(4>=4)

print(4==4)

print(4!=4)
````

````python
# Logical Operator
print(1 and 0)

print(1 or 0)

print(not 1)
````

````python
# Bitwise Operators

# bitwise and---------------------------------
print(2 & 3)     # 010
                 # 011
                 #------ (and opration)
                 # 010  = 2
# bitwise or----------------------------------
print(2 | 3)    # 010
                # 011
                #------- (or operation)
                # 011  = 3
# bitwise xor---------------------------------
print(2 ^ 3)   # 010
               # 011
               #--------- (xor operation)
               # 001  = 1
# bitwise not---------------------------------
print(~3)     # 3 in binary (8-bit for example):
              # 00000011
              # Applying bitwise NOT (~) flips all bits:
              # 11111100
              # This binary pattern represents -4 in two’s complement.
              # -4 (Output)
              #-------------------------------
              #  Simple Rule:
              #  ~n = -(n + 1)
              #  ~3 = -(3 + 1) = -4

# Right shift operator------------------------
print(4 >> 2)            # 1) Start value (decimal): 4
                         # 2) Binary of 4: 100
                         # 3) Operation: right shift by 2 (>> 2) — remove 2 least-significant bits
                         #    100 >> 2 => 1
                         # 4) Binary 1 in decimal is 1
                         # result: 1







````

````python
# Assignment Operators ( = )

a = 2
print(a)
a += 2
print(a)
a %= 2
print(a)
````

````python
# Membership Opeator ( in / not in)
print('D' in 'Delhi')
print('M' not in 'Muzaffarpur')

print(1 in [2,3,4,5,6])
print(2 in [5,4,2,6,7])
````
