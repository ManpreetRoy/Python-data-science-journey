##### Q. Wrtie a program to enter two integer and then perform all arithmetic operation on them ?

````python
x = int(input("Enter first number : "))
y = int(input("Enter second number : "))

print("Add : ",x+y)
print("Sub : ",x-y)
print("Mul : ",x*y)
print("Div : ",x/y)
print("Mod : ",x%y)
print("Expo : ",x**y)
````

#### Q. Write a program to print the ASCII value of a charcter ?

````python
print("ASCII Value | Character")
print("-----------------------")
# Iterate through all standard ASCII values (0 to 127)
for i in range(128):
    # Get the character corresponding to the ASCII value
    char = chr(i)
    # Print the value and the character
    print(f"{i:<12}| {char}")

````
